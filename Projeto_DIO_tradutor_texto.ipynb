```python
import requests
from docx import Document
import os

subscription_key = "CFORapdZRwV1xq0jkCJe1jltRhxdkwpfluM8G8HhtDYq4hGY3QvPJQQJ99CAACULyCpXJ3w3AAAbACOGfvhc" # Replace with your actual subscription key
endpoint = 'https://api.cognitive.microsofttranslator.com/'
location = "Global"
language_destination = 'pt-br'

def translator_text(text, target_language):
    path = '/translate'
    constructed_url = endpoint + path
    headers = {
        'Ocp-Apim-Subscription-Key': subscription_key,
        'Ocp-Apim-Subscription-Region': location,
        'Content-type': 'application/json',
        'X-ClientTraceId': str(os.urandom(16))
    }

    body = [{
        'text': text
    }]

    params = {
        'api-version': '3.0',
        'from': 'en',
        'to': target_language
    }

    request = requests.post(constructed_url, params=params, headers=headers, json=body)

    if request.status_code != 200:
        print(f"API request failed with status code: {request.status_code}")
        print(f"Response body: {request.text}")
        raise Exception(f"Translator API error: {request.status_code} - {request.text}")

    response = request.json()
    if not response or "translations" not in response[0] or not response[0]["translations"]:
        print(f"Unexpected API response structure: {response}")
        raise Exception("Unexpected API response structure or missing translation data.")

    return response[0]["translations"][0]["text"]
```

```python
def translator_document(path):
  document = Document(path)
  full_text = []
  for paragraph in document.paragraphs:
      translated_text = translator_text(paragraph.text, language_destination)
      full_text.append(translated_text)

  translated_doc = Document()
  for line in full_text:
    translated_doc.add_paragraph(line)
  path_translated = path.replace(".docx", f"_{language_destination}.docx")
  translated_doc.save(path_translated)

  return full_text
```
